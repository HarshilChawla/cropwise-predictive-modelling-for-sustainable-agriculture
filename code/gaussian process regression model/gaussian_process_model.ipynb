```python
#For making datasets and feature files - DO NOT run again and again

import os
from sklearn.feature_selection import SelectKBest, f_regression

dataset_folder = "model_datasets/"
os.makedirs(dataset_folder, exist_ok=True)

for crop, state in valid_pairs_final:
    df_pair = df[(df["Crop"] == crop) & (df["State"] == state)].copy()
    if len(df_pair) < 10:
        continue

    y = df_pair["Yield (t/ha)"]
    X = df_pair.drop(["Yield (t/ha)", "Crop", "State", "Production (tonnes)"], axis=1)
    X_encoded = pd.get_dummies(X, columns=["Season"], drop_first=True)

    feature_cols_no_year = [c for c in X_encoded.columns if c != "Year"]
    X_no_year = X_encoded[feature_cols_no_year]

    k_features = min(25, X_no_year.shape[1])
    selector = SelectKBest(score_func=f_regression, k=k_features)
    X_selected_no_year = selector.fit_transform(X_no_year, y)
    selected_cols_no_year = X_no_year.columns[selector.get_support(indices=True)]

    selected_cols = ["Year"] + list(selected_cols_no_year)

    df_final = pd.DataFrame(X_encoded[selected_cols], columns=selected_cols)
    features_folder = "selected_features/"
    os.makedirs(features_folder, exist_ok=True)

    feature_file = f"{crop}_{state}_features.txt".replace(" ", "_").replace("/", "-")
    feature_path = os.path.join(features_folder, feature_file)
    with open(feature_path, "w") as f:
        f.write("\n".join(selected_cols))

    df_final["Yield (t/ha)"] = y.values
    file_name = f"{crop}_{state}.csv".replace(" ", "_").replace("/", "-")
    save_path = os.path.join(dataset_folder, file_name)

    df_final.to_csv(save_path, index=False)
```

```python
#Gaussian Process Regression
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

dataset_folder = "model_datasets/"
plot_folder = "plots/"
os.makedirs(plot_folder, exist_ok=True)

metrics_list = []

for file in os.listdir(dataset_folder):
    if not file.endswith(".csv"):
        continue

    path = os.path.join(dataset_folder, file)
    df_model = pd.read_csv(path)

    train_df = df_model[df_model["Year"] < 2016]
    test_df = df_model[df_model["Year"] > 2015]

    if len(test_df) == 0:
        continue

    X_train = train_df.drop("Yield (t/ha)", axis=1)
    y_train = train_df["Yield (t/ha)"]

    X_test = test_df.drop("Yield (t/ha)", axis=1)
    y_test = test_df["Yield (t/ha)"]

    kernel = RBF(length_scale=1.0) + WhiteKernel(noise_level=0.1)
    model = GaussianProcessRegressor(kernel=kernel, normalize_y=True)

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    r2 = r2_score(y_test, preds)

    metrics_list.append([file.replace(".csv", ""), mae, rmse, r2])

    plt.figure(figsize=(6, 6))
    plt.scatter(y_test, preds, s=40)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
    plt.xlabel("Actual Yield")
    plt.ylabel("Predicted Yield")
    plt.title(file.replace(".csv", ""))

    plot_path = os.path.join(plot_folder, file.replace(".csv", "") + ".png")
    plt.savefig(plot_path, dpi=300)
    plt.close()

metrics_df = pd.DataFrame(metrics_list, columns=["Model", "MAE", "RMSE", "R2"])
metrics_df.to_csv("model_metrics.csv", index=False)

print(metrics_df)
```

```python
import os
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# CONFIG
dataset_folder = "model_datasets/"
features_folder = "selected_features/"
plot_folder = "plots/"
pkl_folder = "model_pkl_file/"   
os.makedirs(plot_folder, exist_ok=True)
os.makedirs(dataset_folder, exist_ok=True)
os.makedirs(features_folder, exist_ok=True)

SAVE_PKLS = True  
PKL_R2_THRESHOLD = -0.10   
PKL_REL_RMSE_PERCENT = 40.0  

final_pairs = [p for p in valid_pairs if p not in pairs_to_drop]

metrics = []
skipped = []
trained_count = 0
saved_count = 0

for crop, state in final_pairs:
    model_name = f"{crop}_{state}".replace(" ", "_").replace("/", "-")
    csv_file = os.path.join(dataset_folder, model_name + ".csv")
    feat_file = os.path.join(features_folder, model_name + "_features.txt")

    if not os.path.exists(csv_file):
        skipped.append((model_name, "no_csv"))
        continue
    if not os.path.exists(feat_file):
        skipped.append((model_name, "no_features"))
        continue

    df_model = pd.read_csv(csv_file)

    with open(feat_file, "r") as f:
        features = [ln.strip() for ln in f.readlines() if ln.strip()]

    if "Year" not in features:
        features = ["Year"] + features

    for col in features:
        if col not in df_model.columns:
            df_model[col] = 0.0

    df_model = df_model[features + ["Yield (t/ha)"]]

    train_df = df_model[(df_model["Year"] >= 1997) & (df_model["Year"] <= 2015)].copy()
    test_df = df_model[(df_model["Year"] >= 2016) & (df_model["Year"] <= 2019)].copy()

    if len(train_df) < 5:
        skipped.append((model_name, "too_few_train"))
        continue
    if len(test_df) < 2:
        skipped.append((model_name, "too_few_test"))
        continue

    X_train = train_df.drop("Yield (t/ha)", axis=1).astype(float)
    y_train = train_df["Yield (t/ha)"].astype(float)

    X_test = test_df.drop("Yield (t/ha)", axis=1).astype(float)
    y_test = test_df["Yield (t/ha)"].astype(float)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    kernel = ConstantKernel(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e3)) \
             + WhiteKernel(noise_level=1.0, noise_level_bounds=(1e-6, 1e6))

    model = GaussianProcessRegressor(kernel=kernel, normalize_y=True, n_restarts_optimizer=5)

    model.fit(X_train_scaled, y_train)
    preds, sigma = model.predict(X_test_scaled, return_std=True)

    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    r2 = r2_score(y_test, preds)
    rel_rmse_pct = (rmse / max(1e-8, np.mean(y_train))) * 100.0

    metrics.append([model_name, mae, rmse, r2, rel_rmse_pct, len(train_df), len(test_df)])

    trained_count += 1

    plt.figure(figsize=(6, 5))
    years = test_df["Year"].values
    order = np.argsort(years)
    years_sorted = years[order]
    y_test_sorted = y_test.values[order]
    preds_sorted = preds[order]
    sigma_sorted = sigma[order]

    plt.plot(years_sorted, y_test_sorted, marker="o", label="Actual")
    plt.plot(years_sorted, preds_sorted, marker="o", label="Predicted")
    plt.fill_between(years_sorted,
                     preds_sorted - 1.96 * sigma_sorted,
                     preds_sorted + 1.96 * sigma_sorted,
                     alpha=0.2, label="95% CI")
    plt.xlabel("Year")
    plt.ylabel("Yield (t/ha)")
    plt.title(model_name)
    plt.legend()
    plt.tight_layout()
    plot_path = os.path.join(plot_folder, model_name + ".png")
    plt.savefig(plot_path, dpi=200)
    plt.close()

    if SAVE_PKLS:
        if (r2 >= PKL_R2_THRESHOLD) and (rel_rmse_pct <= PKL_REL_RMSE_PERCENT):
            os.makedirs(pkl_folder, exist_ok=True)
            save_obj = {
                "model": model,
                "scaler": scaler,
                "features": features
            }
            joblib.dump(save_obj, os.path.join(pkl_folder, model_name + ".pkl"))
            saved_count += 1

metrics_df = pd.DataFrame(metrics, columns=["Model", "MAE", "RMSE", "R2", "Rel_RMSE_pct", "Train_rows", "Test_rows"])
metrics_df.to_csv("model_metrics_fixed.csv", index=False)

print(f"Trained models: {trained_count}, Saved models: {saved_count}, Skipped models: {len(skipped)}")
print("Top 10 worst by Rel_RMSE_pct")
print(metrics_df.sort_values("Rel_RMSE_pct", ascending=False).head(10))
print("Top 10 best by R2")
print(metrics_df.sort_values("R2", ascending=False).head(10))
```

```python
import pandas as pd
import os
metrics = pd.read_csv("model_metrics_fixed.csv")
metrics["Crop"] = metrics["Model"].apply(lambda x: x.split("_")[0])
metrics["State"] = metrics["Model"].apply(lambda x: "_".join(x.split("_")[1:]))
avg_yields = []
dataset_folder = "model_datasets/"
for model_name in metrics["Model"]:
    csv_file = os.path.join(dataset_folder, model_name + ".csv")
    if os.path.exists(csv_file):
        df_temp = pd.read_csv(csv_file)
        avg_yields.append(df_temp["Yield (t/ha)"].mean())
    else:
        avg_yields.append(None)

metrics["AvgYield"] = avg_yields
metrics["Rel_RMSE_pct"] = (metrics["RMSE"] / metrics["AvgYield"]) * 100

good_models = metrics[
    (metrics["R2"] > 0) &
    (metrics["Rel_RMSE_pct"] < 25) &
    (metrics["RMSE"] < (0.30 * metrics["AvgYield"])) &
    (metrics["MAE"] < (0.25 * metrics["AvgYield"]))
].copy()

good_models.to_csv("final_selected_models.csv", index=False)
print("Total Models:", len(metrics))
print("Selected Good Models:", len(good_models))
print("Dropped Models:", len(metrics) - len(good_models))
print("\nTop 10 Good Models:")
print(good_models[["Model", "MAE", "RMSE", "R2", "Rel_RMSE_pct"]].head(10))
```
